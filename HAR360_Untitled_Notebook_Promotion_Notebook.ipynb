# HAR360 Untitled Notebook Promotion Notebook

Promotes high-value `Untitled*.ipynb` notebooks discovered by the Untitled Notebook Scanner into `HAR360_MASTER/canonical_staging` as evidence-backed recovery assets.

```python
# ===============================
# CELL 1 — MOUNT GOOGLE DRIVE
# ===============================

from google.colab import drive
drive.mount('/content/drive')
```

```python
# ===============================
# CELL 2 — IMPORTS
# ===============================

from pathlib import Path
from datetime import datetime
import json
import shutil
import hashlib
import pandas as pd
```

```python
# ===============================
# CELL 3 — CONFIGURATION
# ===============================

MASTER_ROOT = Path('/content/drive/MyDrive/HAR360_MASTER')
UNTITLED_SCAN_ROOT = MASTER_ROOT / 'untitled_notebook_scan'
UNTITLED_REPORTS_ROOT = UNTITLED_SCAN_ROOT / 'reports'
UNTITLED_CLASSIFIED_ROOT = UNTITLED_SCAN_ROOT / 'classified'

CANONICAL_STAGING_ROOT = MASTER_ROOT / 'canonical_staging'
PROMOTION_ROOT = CANONICAL_STAGING_ROOT / '09_untitled_recovery'
PROMOTION_REPORTS_ROOT = PROMOTION_ROOT / 'reports'

for folder in [PROMOTION_ROOT, PROMOTION_REPORTS_ROOT]:
    folder.mkdir(parents=True, exist_ok=True)

# Minimum score for automatic promotion. Lower if needed.
MIN_PROMOTION_SCORE = 150.0

# Categories worth promoting into canonical staging.
PROMOTE_CATEGORIES = {
    'canonical_reconstruction',
    'core_runtime',
    'registry_tools',
    'trainer_ai',
    'tln_timeline',
    'guardian_codex',
    'economics_finance',
}

print('✅ Configuration ready')
print('Untitled scan root:', UNTITLED_SCAN_ROOT)
print('Promotion root:', PROMOTION_ROOT)
```

```python
# ===============================
# CELL 4 — HELPERS
# ===============================

def sha256_file(path: Path) -> str:
    h = hashlib.sha256()
    with path.open('rb') as f:
        for chunk in iter(lambda: f.read(1024 * 1024), b''):
            h.update(chunk)
    return h.hexdigest()

def safe_name(name: str) -> str:
    keep = []
    for ch in str(name):
        if ch.isalnum() or ch in ('-', '_', '.', ' '):
            keep.append(ch)
        else:
            keep.append('_')
    return ''.join(keep).strip().replace(' ', '_')

def latest_registry_csv() -> Path | None:
    files = sorted(
        UNTITLED_REPORTS_ROOT.glob('UNTITLED_NOTEBOOK_REGISTRY_*.csv'),
        key=lambda p: p.stat().st_mtime,
        reverse=True
    )
    return files[0] if files else None

def resolve_source_path(row) -> Path | None:
    # Prefer the original path from the registry.
    original = Path(str(row.get('path', '')))
    if original.exists():
        return original

    # Fall back to classified copy if available.
    suggested = str(row.get('suggested_name', ''))
    category = str(row.get('category', ''))
    if suggested and category:
        candidate = UNTITLED_CLASSIFIED_ROOT / category / suggested
        if candidate.exists():
            return candidate

    # Search classified tree for the original name.
    name = str(row.get('name', ''))
    if name:
        matches = list(UNTITLED_CLASSIFIED_ROOT.rglob(name))
        if matches:
            return matches[0]

    return None

print('✅ Helpers loaded')
```

```python
# ===============================
# CELL 5 — LOAD LATEST UNTITLED REGISTRY
# ===============================

registry_file = latest_registry_csv()

if registry_file is None:
    raise FileNotFoundError(
        f'No UNTITLED_NOTEBOOK_REGISTRY_*.csv found in {UNTITLED_REPORTS_ROOT}. Run the scanner notebook first.'
    )

records_df = pd.read_csv(registry_file)

print('✅ Loaded registry:')
print(registry_file)
print('Records:', len(records_df))

if not records_df.empty:
    display(records_df.head(20))
```

```python
# ===============================
# CELL 6 — SELECT PROMOTION CANDIDATES
# ===============================

if records_df.empty:
    selected_df = records_df.copy()
else:
    selected_df = records_df[
        (records_df['harmony_score'].astype(float) >= MIN_PROMOTION_SCORE)
        & (records_df['category'].isin(PROMOTE_CATEGORIES))
    ].copy()

selected_df = selected_df.sort_values('harmony_score', ascending=False)

print('Promotion candidates:', len(selected_df))

if not selected_df.empty:
    display(selected_df[['name', 'category', 'harmony_score', 'suggested_name', 'path']].head(50))
```

```python
# ===============================
# CELL 7 — PROMOTE NOTEBOOKS
# ===============================

promotion_records = []
errors = []
seen_hashes = {}

for _, row in selected_df.iterrows():
    try:
        src = resolve_source_path(row)

        if src is None or not src.exists():
            errors.append({
                'name': row.get('name', ''),
                'path': row.get('path', ''),
                'error': 'source_not_found'
            })
            continue

        file_hash = sha256_file(src)
        category = safe_name(row.get('category', 'review_needed'))
        suggested = safe_name(row.get('suggested_name', src.name))

        dest_dir = PROMOTION_ROOT / category
        dest_dir.mkdir(parents=True, exist_ok=True)
        dest = dest_dir / suggested

        # Avoid overwriting and keep duplicates visible.
        if file_hash in seen_hashes:
            duplicate_of = seen_hashes[file_hash]
            action = 'duplicate_skipped'
        else:
            duplicate_of = ''
            if dest.exists():
                stem = dest.stem
                suffix = dest.suffix
                dest = dest_dir / f'{stem}__promoted_{len(promotion_records)+1}{suffix}'
            shutil.copy2(src, dest)
            seen_hashes[file_hash] = str(dest)
            action = 'copied'

        promotion_records.append({
            'original_name': row.get('name', ''),
            'original_path': str(src),
            'promoted_path': str(dest) if action == 'copied' else '',
            'category': category,
            'harmony_score': float(row.get('harmony_score', 0)),
            'suggested_name': suggested,
            'sha256': file_hash,
            'action': action,
            'duplicate_of': duplicate_of,
        })

    except Exception as e:
        errors.append({
            'name': row.get('name', ''),
            'path': row.get('path', ''),
            'error': repr(e)
        })

promotion_df = pd.DataFrame(promotion_records)
errors_df = pd.DataFrame(errors)

print('Promotion records:', len(promotion_df))
print('Errors:', len(errors_df))

if not promotion_df.empty:
    display(promotion_df.head(50))
if not errors_df.empty:
    display(errors_df)
```

```python

# ===============================
# CELL 8 — EXPORT PROMOTION REPORTS
# ===============================

from datetime import datetime
import json

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

PROMOTION_ROOT.mkdir(parents=True, exist_ok=True)
PROMOTION_REPORTS_ROOT.mkdir(parents=True, exist_ok=True)

promotion_csv = PROMOTION_REPORTS_ROOT / f"UNTITLED_PROMOTION_REGISTRY_{timestamp}.csv"
promotion_json = PROMOTION_REPORTS_ROOT / f"UNTITLED_PROMOTION_REGISTRY_{timestamp}.json"
errors_csv = PROMOTION_REPORTS_ROOT / f"UNTITLED_PROMOTION_ERRORS_{timestamp}.csv"
report_md = PROMOTION_REPORTS_ROOT / f"UNTITLED_PROMOTION_REPORT_{timestamp}.md"
har360_report = PROMOTION_REPORTS_ROOT / f"untitled_promotion_{timestamp}.har360.json"

promotion_df.to_csv(promotion_csv, index=False)
promotion_df.to_json(promotion_json, orient="records", indent=2)
errors_df.to_csv(errors_csv, index=False)

category_counts = (
    promotion_df["category"].value_counts().to_dict()
    if not promotion_df.empty and "category" in promotion_df.columns
    else {}
)

action_counts = (
    promotion_df["action"].value_counts().to_dict()
    if not promotion_df.empty and "action" in promotion_df.columns
    else {}
)

lines = []
lines.append("# HAR360 Untitled Notebook Promotion Report")
lines.append("")
lines.append(f"- Created: {timestamp}")
lines.append(f"- Source registry: {registry_file}")
lines.append(f"- Selected candidates: {len(selected_df)}")
lines.append(f"- Promotion records: {len(promotion_df)}")
lines.append(f"- Errors: {len(errors_df)}")
lines.append("")
lines.append("## Category Overview")

for cat, count in category_counts.items():
    lines.append(f"- {cat}: {count}")

lines.append("")
lines.append("## Action Overview")

for action, count in action_counts.items():
    lines.append(f"- {action}: {count}")

lines.append("")
lines.append("## Top Promoted Untitled Notebooks")

if not promotion_df.empty:
    sorted_df = promotion_df.sort_values(
        "harmony_score",
        ascending=False
    )

    for _, row in sorted_df.head(30).iterrows():
        lines.append(
            f"- `{row.get('original_name', '')}` "
            f":: score={float(row.get('harmony_score', 0)):.1f} "
            f":: {row.get('category', '')} "
            f":: action={row.get('action', '')} "
            f":: `{row.get('promoted_path', '')}`"
        )

lines.append("")
lines.append("## Next Action")
lines.append(
    "Inspect promoted notebooks and include the highest-value ones "
    "in Harmony360 class/runtime merger evidence."
)

report_md.write_text(
    "\n".join(lines),
    encoding="utf-8"
)

snapshot = {
    "snapshot_type": "HAR360_UNTITLED_NOTEBOOK_PROMOTION",
    "created_at": timestamp,
    "source_registry": str(registry_file),
    "promotion_root": str(PROMOTION_ROOT),
    "reports_root": str(PROMOTION_REPORTS_ROOT),
    "selected_count": int(len(selected_df)),
    "promotion_record_count": int(len(promotion_df)),
    "error_count": int(len(errors_df)),
    "category_counts": category_counts,
    "action_counts": action_counts,
    "records": (
        promotion_df.to_dict(orient="records")
        if not promotion_df.empty
        else []
    ),
}

har360_report.write_text(
    json.dumps(snapshot, indent=2, ensure_ascii=False),
    encoding="utf-8"
)

print("✅ Untitled notebook promotion exported")
print("Promotion CSV:", promotion_csv)
print("Promotion JSON:", promotion_json)
print("Errors CSV:", errors_csv)
print("Report MD:", report_md)
print("HAR360 Report:", har360_report)
```

```python
# ===============================
# CELL 9 — FINAL SUMMARY
# ===============================

print()
print('===============================')
print('HAR360 UNTITLED PROMOTION COMPLETE')
print('===============================')
print(f'Selected candidates : {len(selected_df)}')
print(f'Promotion records   : {len(promotion_df)}')
print(f'Errors              : {len(errors_df)}')

print()
print('Promotion folder:')
print(PROMOTION_ROOT)

print()
print('Reports folder:')
print(PROMOTION_REPORTS_ROOT)

print()
print('Top promoted:')
if not promotion_df.empty:
    for _, row in promotion_df.sort_values('harmony_score', ascending=False).head(15).iterrows():
        print(
            f"- {row.get('original_name', '')} "
            f":: score={float(row.get('harmony_score', 0)):.1f} "
            f":: {row.get('category', '')} "
            f":: {row.get('action', '')}"
        )
else:
    print('No notebooks promoted.')
```
